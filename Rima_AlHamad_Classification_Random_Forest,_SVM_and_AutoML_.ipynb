```python
#Prepare the data for analysis

import pandas as pd
import numpy as np

data = pd.read_table('https://raw.githubusercontent.com/PineBiotech/omicslogic/master/LIHC_RSEM_42cases.txt',sep='\t',header=(0))

features = data.iloc[:, 0].values

#len(data.columns)
data = data.T.reset_index()
#grab the first row for the header
new_header = data.iloc[0] 
#take the data less the header row
data = data[1:] 
data.columns = new_header
data
```

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/PineBiotech/omicslogic/master/TCGA_43_clinical.csv',header=(0))
# taking into account only the first 16 characters of the string
data2['Variable']=data2['Variable'].str[:16]  
# replacing the - with _ to match the data in data
data2['Variable'] = data2['Variable'].str.replace('-','_')  

data2 = data2[['Variable','race']]

data2
```

```python
df = data2.merge(data, right_on='ID', left_on='Variable', how='right').drop(columns=['Variable','ID'])
df
```

```python
# number of unique values in the target column
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
classes = np.unique(y)
len(classes)  
```

```python
str(classes)  # list of unique columns in the target column
```

SMV

```python
# initially SVM is a binary (two classes) classifier
# so by default multiclass problems are handled according to a "one-vs-one" scheme.
# if you want to try OneVsRest scheme - you can use OneVsRestClassifier from sklearn as a wrapper around SVC.
# import model class (SVC stands for Support Vector Classification)
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.preprocessing import scale

X = scale(X, axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# there are several useful parameters such as:
# C - Regularization parameter. The strength of the regularization.
# kernel - Specifies the kernel type to be used in the algorithm
# we will use a linear kernel, it is the only kernel which allows us to use features selection
model = LinearSVC(C=1.0)
```

```python
# now we can easily train our model on the previously loaded data
model.fit(X_train, y_train)

# after training we can predict labels for the test samples
predictions = model.predict(X_test)
predictions
```
